```python
class Nodo:
    def __init__(self, value, siguiente = None):
        self.data = value
        self.next = siguiente
```

```python
head = Nodo(10)
print(head.data)           #Se obtuvo el nodo con valor 10.
```

```python
head = Nodo(10)
head.next = Nodo(20)
print(f"Primero = {head.data}")         #Se obtuvo el nodo con valor 10 y 20.
print(f"Segundo = {head.next.data}")

#Lo mismo pero ahorrando codigo
head = Nodo(10, Nodo (20))
print(f"Primero = {head.data}")
print(f"Segundo = {head.next.data}")

#Transversal

curr_node = head
while curr_node != None:
        print(f"| {curr_node.data} | -->", end="")
        curr_node = curr_node.next
print("")
```

```python
#Insertar valores
curr_node = head
while curr_node.data != 20: #Buscamos el nodo
    curr_node = curr_node.next
print(curr_node.data) #Corrobar que estamos en el ultimo

#   Insertar el 25 entre 20 y 30
curr_node = head
while curr_node.data != 20:
    curr_node = curr_node.next
print(curr_node.data)

#      Crear Nodo 25
curr_node.next = Nodo(25, curr_node.next )
print("Después de insertar a 25 ")
#   Trasnversal
curr_node = head
while curr_node !=  None:
    print(f"| {curr_node.data } | ---> " , end="")
    curr_node = curr_node.next
print("")
```

```python
class Nodo:
  def __init__(self, value, siguiente = None):
    self.data = value
    self.next = siguiente

class LinkedList:
  def __init__(self):
    self.head = None
  
  def is_empty(self):
    return self.head == None

  def get_tail(self):
    curr_node = self.head
    if self.head != None:
      while curr_node.next != None:
        curr_node = curr_node.next
    return curr_node

  def append(self, dato):
    if self.is_empty():
      self.head = Nodo(dato)
    else:
      self.get_tail().next = Nodo(dato)

  def prepend(self, dato):
    if self.is_empty():
      self.head = Nodo(dato)
    else:
      self.head = Nodo(dato,self.head)

  def remove(self, dato):
    curr_node = self.head
    aux = False
    self.head = None
    while curr_node != None:
      if curr_node.data == dato:
        aux = True
        curr_node = curr_node.next
        break   
 
      self.append(curr_node.data)  
      curr_node = curr_node.next

    while curr_node != None:                 
      self.append(curr_node.data)  
      curr_node = curr_node.next               

    if aux == False:
      print("Elemento no encontrado")

  def transversal(self):
    curr_node = self.head
    while curr_node != None:
      print(curr_node.data, end=" --> ")
      curr_node = curr_node.next
    print()
   
  def pop(self):
    curr_node = self.head
    self.head = None
    while curr_node != None:
      if curr_node.next ==  None:
        return curr_node         
      else:
        self.append(curr_node.data)
        curr_node = curr_node.next

  def pop_first(self):
    tmp = self.head             
    curr_node = self.head.next
    self.head = None
    while curr_node != None:
      self.append(curr_node.data)
      curr_node = curr_node.next
    return tmp

  def add_after(self, reference_value, dato):
    curr_node = self.head
    self.head = None
    while curr_node != None:
      if curr_node.data == reference_value:
        self.append(curr_node.data)
        self.append(dato)
        curr_node = curr_node.next
      else:
        self.append(curr_node.data)  
        curr_node = curr_node.next
```

```python
lista = LinkedList()
lista.append(10)
lista.append(15)
lista.append(20)
lista.append(25)
print("------ Lista inicial ------")
lista.transversal()
print("----- Insertando nodo 23 -----")
lista.add_after(20,23)
lista.transversal()
print("----- Eliminando nodo 25 -----")
lista.remove(25)
lista.transversal()
```

#Lista doblemente ligada (DoubleLinkedList).


```python
#Crearemos el nodo doble
class NodoDoble:
    def __init__(self, value, siguiente = None, previo = None):
        self.data = value
        self.next = siguiente
        self.prev = previo
```

- Crear un nodo (10)
    - Imprimir el valor del nodo
- Crear dos nodos (10, 20)
    - Imprimir el valor del primero
    - Imprimir el valor del segundo
    - Transversal
- Lo mismo pero con cuatro nodos (10, 20, 30, 40)
    - Imprimir el valor del primero
    - Imprimir el valor del ultimo
    - Transversal
    - Reverse transversal
    - Insertar el nodo final


Todo esto antes de ir a la escritura formal de ADT Lista Ligada

```python
#Ejercicio 1
print("Ejercicio 1")
head = NodoDoble(10)
tail = head
print(head.data)
print(tail.data) #Otra manera de imprimir el NODO puesto que TAIL hace referencia a HEAD y HEAD tien el valor del nodo

#Ejercicio 2
print("Ejercicio 2")
head = NodoDoble(10, NodoDoble(20))
head.next.prev = head
tail = head.next
print("El primero: ", head.data)
print("El primero ", tail.prev.data) #Otra manera de imprimir el NODO.

print("El SEGUNDO: ", head.next.data)
print("El SEGUNDO ", tail.data)

#Ejercico 3

print( "Ejercicio 3" )
head = NodoDoble ( 10, NodoDoble (20, NodoDoble (30, NodoDoble(40))))
head.next.prev = head
head.next.next.prev = head.next
tail.prev = head.next.next
print("el primero:" , head.data ) 
print("el ultimo es:" , tail.prev.prev.prev.data )

#Transversal
curr_node = head
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.next
print("")

#Reverse transversal
curr_node = tail
while curr_node != None:
    print(f"|{curr_node.data}|-->", end="")
    curr_node = curr_node.prev
print("")

# agregar elemento al final
print("Agregando 50 al final")
tail.next = NodoDoble( 50, None, tail)

# Transversal
curr_node = head
while curr_node != None:
    print(f" {curr_node.data } |  --> " , end="")
    curr_node = curr_node.next
print("")
```

```python
class DoubleLinkedList:
    def __init__( self ):
        self.head = None
        self.tail = None
        self.size = 0

    def get_size( self ):
        return self.size
    
    def is_empty( self ): # true en todo caso de que la lista este vacia.
        return self.size == 0
    
    def append( self , value ):
        if self.is_empty():
            self.head = NodoDoble( value )
            self.tail = self.head
        else:
            self.tail.next = NodoDoble( value , None , self.tail )
            self.tail = self.tail.next
        self.size += 1
    
    def find_from_tail( self , value ):
        curr_node = self.tail
        contador = 0
        while curr_node.data != value:
            contador += 1
            curr_node = curr_node.prev
        
        if curr_node.data == value:
            print(f"Su Nodo está en la posición: { contador + 1 } entrando por tail")

    def find_from_head( self , value ):
        curr_node = self.head
        contador = 0
        while curr_node.data != value:
            contador += 1
            curr_node = curr_node.next

        if curr_node.data == value:
            print(f"Su Nodo está en la posición: { contador + 1 } entrando por head")

    def remove_from_tail( self , value ):
        curr_node = self.tail
        while curr_node.data != value:
            curr_node = curr_node.prev
        
        if curr_node.data == value:
            curr_node.next.prev = curr_node.prev
            curr_node.prev.next = curr_node.next
        #self.size += 1

    def remove_from_head( self , value ):
        curr_node = self.head
        while curr_node.data != value:
            curr_node = curr_node.next

        if curr_node.data == value:
            curr_node.prev.next = curr_node.next
            curr_node.next.prev = curr_node.prev
        #self.size -= 1

    def transversal( self ):
        curr_node = self.head
        while curr_node != None:
            print(f"| {curr_node.data} | -->", end="")
            curr_node = curr_node.next
        print("")


    def reverse_transversal( self ):
        curr_node = self.tail
        while curr_node != None:
            print(f"| {curr_node.data} | -->", end="")
            curr_node = curr_node.prev
        print("")
```

```python
ldl = DoubleLinkedList()
ldl.append(10)
ldl.append(20)
ldl.append(30)
ldl.append(40)
ldl.append(50)
ldl.append(60)
ldl.transversal()
print("")
print(f"¿La lista está vacía?: { ldl.is_empty() }")
print("")
print("Después de remove_from_head")
ldl.remove_from_head(20)
ldl.transversal()
print("")
print("Después de remove_from_tail")
ldl.remove_from_tail(30)
ldl.transversal()
print("")
print(f"La Lista Doblemente Ligada tiene { ldl.get_size() } elementos")
print("")
ldl.find_from_tail(10)
print("")
ldl.find_from_head(50)
print("")
print("reverse_transversal:")
ldl.reverse_transversal()
print("")
```
